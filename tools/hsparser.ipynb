```python
import re
import numpy as np

def parseStringToCodes(string):
    """
    Expand a string of contracted HS codes to a dict of length 2 that contains
    a positive list ('pos') and a negative list ('neg').
    """
    # Strip whitespaces
    string = re.sub(' ', '', string)
    # Capture positive and negative lists
    pos = re.sub('減\((.*?)\)', '', string)
    neg = ';'.join(re.findall('減\((.*?)\)', string))
    
    def expandRange(matchObject):
        """
        Expand codes of the form "XXXX~XXXX" to "XXXX,...,XXXX".
        """
        return ','.join([str(x).zfill(len(matchObject.group(1)))
                         for x in range(int(matchObject.group(1)),
                                        int(matchObject.group(2)) + 1)])
        
    def appendPrefix(matchObject):
        """
        Convert "XXXX-YYYY,...,YYYY" to "XXXX-YYYY;...;XXXX-YYYY".
        """
        prefix = matchObject.group(1) + '-'
        ls = [prefix + s for s in matchObject.group(2).split(',')]
        return ';'.join(ls)
    
    def parseString(string):
        
        # Capture tildes and apply expandRange()
        string = re.sub('(\d+)~(\d+)', expandRange, string)
        # Capture dashes and apply appendPrefix()
        string = re.sub('(\d+)-([^;]*)', appendPrefix, string)
        # Remove all dashes
        string = re.sub('-', '', string)
        # Create a list of codes
        return re.split(';|,', string)
    
    codes = dict(pos=parseString(pos), neg=parseString(neg))
    
    # Verify all codes have length <= 10
    for k, v in codes.items():
        a = np.array(codes[k])
        invalid_idx = (np.array(list(map(len, a))) > 10)
        invalid_codes = a[invalid_idx]
        if len(invalid_codes):
            raise ValueError('Input string returns codes with length > 10: {}'.format(
                    ', '.join(invalid_codes)))
    
    return codes
```

```python
import pandas as pd

def outputCodes(string, filename, ver='2017'):
    """
    Uses parseStringToCodes() to convert the input string to a dict,
    searches the HS code table for matches, and outputs to a text file.
    """
    # Convert input string to dict
    code_dict = parseStringToCodes(string)
    
    # Reference HS codes
    path = '//172.20.23.190/ds/Raw Data/HS code分類表/'
    tens = pd.read_csv(path + ver + '/note_10_C.txt', sep=' ', usecols=['HS_NO'],
                       dtype=str)['HS_NO']
    tens = ','.join(list(tens))
    
    def makeRegex(code_list):
        group = '|'.join([c + '\d*' for c in code_list])
        return '(^|,)(' + group + ')'
    
    # Match positive list against reference codes
    pos_match = re.findall(makeRegex(code_dict['pos']), tens)
    match = [tp[1] for tp in pos_match]
    
    # If negative list non-empty, remove them from matches
    if (len(code_dict['neg']) == 1) and ('' in code_dict['neg']):
        pass
    else:
        neg_match = re.findall(makeRegex(code_dict['neg']), tens)
        neg_match = [tp[1] for tp in neg_match]
        match = [x for x in match if x not in neg_match]
    
    # Write to text file
    with open(filename, 'w') as f:
        for c in match:
            f.write(c + '\n')
    print('{} codes written to {}.'.format(len(match), filename))
    
    return
```
